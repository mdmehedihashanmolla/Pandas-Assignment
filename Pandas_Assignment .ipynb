### 9. Write a Python program to get the top 10 countries data (Last Update, Country/Region, Confirmed, Deaths, Recovered) of Novel Coronavirus (COVID-19).

```python
Covid_Analysing9 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-18-2020.csv', usecols = ['Last Update','Country/Region','Confirmed','Deaths','Recovered'])
```

```python

Sorting_data = Covid_Analysing9.groupby('Country/Region').max().sort_values(by='Confirmed', ascending=False)[:10].reset_index()
pd.set_option('display.max_columns', None)
print(Sorting_data)
```
